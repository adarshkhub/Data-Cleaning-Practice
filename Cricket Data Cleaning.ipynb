```python
import pandas as pd

#read in the csv as a dataframe

df = pd.read_csv("C:/Users/adars/Desktop/Practice_Python_Data/Cricket_Data_1.csv")
df.head()
```

```python
#Rename multiple columns in the list for better readability 

df = df.rename(columns = {"NO" : "Not_Out", "HS" : "Highest_Innings_Scored", "BF" : "Balls_Faced", "SR" : "Strike_Rate"})
```

```python
#Check for Null Values

df.isnull().any()
```

```python
#Drop Duplicates
df.duplicated()

#No duplicates needed to be dropped
```

```python
#Split up span into start and end date, and eventually create a column to display career length
df['Span'].str.split(pat = "-")
```

```python
# Set the start date as the rookie year column
df['Rookie_Year'] = df['Span'].str.split(pat = "-").str[0]
```

```python
df.head()
```

```python
#Set the end date as final year
df['Final_Year'] = df['Span'].str.split(pat = "-").str[0]
```

```python
#Remove the Span Column

df.drop('Span', axis = 1)
```

```python
# Split up the country from the player
df['Country'] = df['Player'].str.split(pat = "(").str[1]
   
#df['Player'].str.split(pat = "(").str[0]
```

```python
df.head()
```

```python
df['Country'] = df['Country'].str.split(pat = ")").str[0]
```

```python
df.head()
```

```python
# show the player column without the country 
df['Player'] = df['Player'].str.split(pat = "(").str[0]
```

```python
df.head()
```

```python
# Changing datatype to make data in column more useable
df.dtypes
```

```python
#Change Highest_Innings_Scored and Balls_Faced to integer type
df.head()

```

```python
df['Balls_Faced'] = df['Balls_Faced'].astype('int64')
```

```python
#Balls_Faced has rows with value "-" which can not be converted to int, so replace these rows with 0 then convert to int

df['Balls_Faced'] = df['Balls_Faced'].replace("-","0")
```

```python
df['Balls_Faced'] = df['Balls_Faced'].astype("int")
```
